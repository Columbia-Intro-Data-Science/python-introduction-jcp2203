```python
import pandas as pd
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import glob
from PIL import Image
from skimage import io, color
import cv2 
%matplotlib inline

imgList1=[];# imgList2=[]; imgList3=[]; imgList4=[]; imgList5=[]; imgList6=[];
labList1=[];
for filename in glob.glob('C:/Users/User/Documents/Class/Data Science/Project_pictures/Final_Images/MelanomaMimic/*.jpg'):
    im = cv2.imread(filename)
    imgList1.append(im) #contains matrix of images
    
nfiles = len(imgList1)
print(nfiles)
```

```python
#check image
rgb = imgList1[36]
plt.imshow(rgb)
#OpenCV uses BGR and Matplotlib uses RGB
RGB_img = cv2.cvtColor(rgb,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)

plt.subplot(1,2,1),plt.imshow(rgb)
plt.title('BGR')
plt.subplot(1,2,2),plt.imshow(RGB_img)
plt.title('RGB')
```

```python
#n, bins, patches = plt.hist(lab[:,:,0], 100, normed=1, facecolor='green', alpha=0.75)
#x = np.linspace(0.0, 30.0, num=1)
lab = color.rgb2lab(RGB_img)
#lab
L = (lab[:,:,0])
a = (lab[:,:,1])
b = (lab[:,:,2])
```

```python
# make black = Nan
l_temp = L
l_temp[l_temp == 0] = float('nan')
L = (l_temp)

a_temp = a
a_temp[a_temp == 0] = float('nan')
a = (a_temp)

b_temp = b
b_temp[b_temp == 0] = float('nan')
b = (b_temp)

#update lab with Nan in black
lab[:,:,0] = L
lab[:,:,1] = a
lab[:,:,2] = b
```

```python
#plot histogram of background-free segmented images
n, bins, patches = plt.hist(L[~np.isnan(L)]) #intensity
n, bins, patches = plt.hist(a[~np.isnan(a)]) #Green-Red
n, bins, patches = plt.hist(b[~np.isnan(b)]) #Blue-Yellow
plt.legend(['(L)Intensity','(A)Green-Red','(B)Blue-Yellow'])
plt.title('L A B Histogram of Segmented Image')
plt.ylabel('Density')
plt.xlabel('L A B')
```

```python
def find_peaks(a):
    x = np.array(a)
    max = np.max(x)
    lenght = len(a)
    ret = []
    for i in range(lenght):
        ispeak = True
        if i-1 > 0:
            ispeak &= (x[i] > 1. * x[i-1])
        if i+1 < lenght:
            ispeak &= (x[i] > 1. * x[i+1])
            ispeak &= (x[i] > 0.05 * max)
        if ispeak:
            ret.append(i)
    return ret
```
