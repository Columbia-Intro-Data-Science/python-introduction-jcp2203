```python
base_dir = r'C:\Users\User\Desktop\TrainTestData'
Xtrain_feat = []
Xtest_feat = []
Xtrain_list=[]
Xtest_list=[]

for i in range(0,len(Xtrain)):
    disease = Xtrain.loc[i][2].split('\\')[-2]
    filename = Xtrain.loc[i][2].split('\\')[-1]
    fullpath = os.path.join(base_dir,disease,filename)
    Xtrain_list.append(fullpath)
    im_temp = mpimg.imread(Xtrain_list[i])
    Xtrain_feat.append(im_temp) 
    

for i in range(0,len(Xtest)):
    disease = Xtest.loc[i][2].split('\\')[-2]
    filename = Xtest.loc[i][2].split('\\')[-1]
    fullpath = os.path.join(base_dir,disease,filename)
    Xtest_list.append(fullpath) 
    im_temp = mpimg.imread(Xtrain_list[i])
    Xtest_feat.append(im_temp)    

#imgloc = Xtrain_list[1]
#mpimg.imread(imgloc)
#mpimg.imread(Xtrain_list(1))
```

```python
Avg_train=[]
Avg_test=[]
peak_index_train=[];
peak_index_test=[]

for i in range(0,len(Xtrain)):
    rgb = Xtrain_feat[i]
    lab = color.rgb2lab(rgb)
    #lab
    L = (lab[:,:,0])
    a = (lab[:,:,1])
    b = (lab[:,:,2])

    # make black = Nan
    l_temp = L
    l_temp[l_temp == 0] = float('nan')
    L = (l_temp)

    a_temp = a
    a_temp[a_temp == 0] = float('nan')
    a = (a_temp)

    b_temp = b
    b_temp[b_temp == 0] = float('nan')
    b = (b_temp)

    #update lab with Nan in black
#    lab[:,:,0] = L
#    lab[:,:,1] = a
#    lab[:,:,2] = b

    Avg_L = np.mean(L[~np.isnan(L)])
    Avg_a = np.mean(a[~np.isnan(a)])
    Avg_b = np.mean(b[~np.isnan(b)])
    
    AVG = [Avg_L,Avg_a,Avg_b]
    Avg_train.append(AVG)
    
    #plot histogram of background-free segmented images
    n1, bins1 = np.histogram(L[~np.isnan(L)]) #intensity
    n2, bins2 = np.histogram(a[~np.isnan(a)]) #Green-Red
    n3, bins3 = np.histogram(b[~np.isnan(b)]) #Blue-Yellow
    
    #Peaks of histogram LAB value
    Pop_L = np.where(n1 == n1.max()) #tuple
    Pop_a = np.where(n2 == n2.max())
    Pop_b = np.where(n3 == n3.max())

    pks=[np.array(Pop_L)[0][0],np.array(Pop_a)[0][0],np.array(Pop_b)[0][0]]
    peak_index_train.append(pks)  
```

```python
for i in range(0,len(Xtest)):
    rgb = Xtest_feat[i]
    lab = color.rgb2lab(rgb)
    #lab
    L = (lab[:,:,0])
    a = (lab[:,:,1])
    b = (lab[:,:,2])

    # make black = Nan
    l_temp = L
    l_temp[l_temp == 0] = float('nan')
    L = (l_temp)

    a_temp = a
    a_temp[a_temp == 0] = float('nan')
    a = (a_temp)

    b_temp = b
    b_temp[b_temp == 0] = float('nan')
    b = (b_temp)

    #update lab with Nan in black
#    lab[:,:,0] = L
#    lab[:,:,1] = a
#    lab[:,:,2] = b

    Avg_L = np.mean(L[~np.isnan(L)])
    Avg_a = np.mean(a[~np.isnan(a)])
    Avg_b = np.mean(b[~np.isnan(b)])
    AVG = [Avg_L,Avg_a,Avg_b]
    Avg_test.append(AVG)
    
    #numpy histogram of background-free segmented images -- prevents wasting time plotting
    n1, bins1 = np.histogram(L[~np.isnan(L)]) #intensity
    n2, bins2 = np.histogram(a[~np.isnan(a)]) #Green-Red
    n3, bins3 = np.histogram(b[~np.isnan(b)]) #Blue-Yellow
    
    #Peaks of histogram LAB value
    Pop_L = np.where(n1 == n1.max()) #tuple
    Pop_a = np.where(n2 == n2.max())
    Pop_b = np.where(n3 == n3.max())

    pks1=[np.array(Pop_L)[0][0],np.array(Pop_a)[0][0],np.array(Pop_b)[0][0]]
    peak_index_test.append(pks1)       
```

```python
#concactenate Avg_train &peak_index_train into one matrix
Xtrain_feat = np.hstack((Avg_train,peak_index_train))#.ravel()
Xtest_feat  = np.hstack((Avg_test,peak_index_test))#.ravel()
#verify peak-index_train is X-value not bin
                         
                         
print(np.atleast_2d(Xtrain_feat).shape[1])
print(len(Xtrain_feat))
                         
print(np.atleast_2d(Xtest_feat).shape[1])
print(len(Xtest_feat))
```

```python
import csv

##produces one row with all of the feature vectors
with open("Xtrain.csv", 'w') as myfile:
     wr = csv.writer(myfile)
     wr.writerow(Xtrain_feat)

with open("Xtest.csv", 'w') as myfile:
     wr = csv.writer(myfile)
     wr.writerow(Xtest_feat)
    
```

```python
#Number of classes 
n = 7

#One hot label encoder
enc = LabelEncoder()
enc.fit(np.arange(1,n+1))
enc.transform(np.arange(1,n+1))
train_labels = enc.transform(np.array(ytrain)[:,1].astype(int))
train_labels = np_utils.to_categorical(train_labels)

print(Xtrain_feat.shape)
print(train_labels)
```

```python
model = Sequential()
#Layer 1
model.add(Dense(10,activation='relu',input_shape=(1,6))) #VERIFY INPUT SHAPE
#model.add(Dense(10,activation='relu',input_dim=6)) #VERIFY INPUT SHAPE
model.add(Dense(n,activation='softmax'))
adam = keras.optimizers.Adam(lr = 1e-4)
model.compile(loss='categorical_crossentropy',
              optimizer=adam,
              metrics=['accuracy'])

batch_size = 32
epoch = 10

filepath = "weights.hdf5"
checkpoint = ModelCheckpoint(filepath,monitor='val_acc',mode='max',
                            save_best_only=True,verbose=1)

tensorboard = TensorBoard(log_dir="./logs3",
                          write_images=True,batch_size=batch_size)
callback_list = [checkpoint,tensorboard]

print(model.summary())
```

```python
H = model.fit(np.expand_dims(Xtrain_feat,axis=0),ytrain,
             batch_size=batch_size,epochs = epoch,verbose=1,
             callbacks=callback_list,validation_split=0.20)
```
