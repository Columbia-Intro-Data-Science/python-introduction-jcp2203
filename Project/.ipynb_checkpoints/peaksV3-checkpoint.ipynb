```python
import pandas as pd
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import glob
from PIL import Image
from skimage import io, color
import cv2 
%matplotlib inline

imgList1=[];# imgList2=[]; imgList3=[]; imgList4=[]; imgList5=[]; imgList6=[];
labList1=[]; 
thresh_L=[]; thresh_a = []; thresh_b=[];
for filename in glob.glob('C:/Users/User/Documents/Class/Data Science/Project_pictures/Final_Images/MelanomaMimic/*.jpg'):
    im = cv2.imread(filename)
    imgList1.append(im) #contains matrix of images
    
nfiles = len(imgList1)
print(nfiles)
```

```python
#check image
rgb = imgList1[36]
plt.imshow(rgb)
#OpenCV uses BGR and Matplotlib uses RGB
RGB_img = cv2.cvtColor(rgb,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)

plt.subplot(1,2,1),plt.imshow(rgb)
plt.title('BGR')
plt.subplot(1,2,2),plt.imshow(RGB_img)
plt.title('RGB')
```

```python
lab = cv2.cvtColor(rgb,cv2.COLOR_BGR2LAB)
L, a, b = cv2.split(lab)
```

```python
#GaussianFiltered Image
#blurL = cv2.GaussianBlur(L,(5,5),0)
#blurA = cv2.GaussianBlur(A,(5,5),0)
#blurB = cv2.GaussianBlur(B.(5,5),0)

#Otsu Thresholding
ret,thresh_L = cv2.threshold(L,70,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret,thresh_a = cv2.threshold(a,70,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret,thresh_b = cv2.threshold(b,70,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
#Plot effects of thresholding
#plt.subplot(1,2,1), 
n, bins, patches = plt.hist(L[~np.isnan(L)]) #intensity
n, bins, patches = plt.hist(a[~np.isnan(a)]) #Green-Red
n, bins, patches = plt.hist(b[~np.isnan(b)]) #Blue-Yellow
plt.legend(['Intensity','Green-Red','Blue-Yellow'])
plt.title('L A B Histogram of Segmented Image')
plt.ylabel('Density')
plt.xlabel('L A B')

```
